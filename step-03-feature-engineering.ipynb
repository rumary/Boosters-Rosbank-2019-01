### Расстояния до ближайших банкоматов

```python
from math import sin, cos, sqrt, atan2, radians

R = 6373.0 # радиус Земли в километрах

def distance(x,y):
    """
    Параметры
    ----------
    x : tuple, широта и долгота первой геокоординаты 
    y : tuple, широта и долгота второй геокоординаты 
    
    Результат
    ----------
    result : дистанция в километрах между двумя геокоординатами
    """
    lat_a, long_a, lat_b, long_b = map(radians, [*x,*y])    
    dlon = long_b - long_a
    dlat = lat_b - lat_a
    a = sin(dlat/2)**2 + cos(lat_a) * cos(lat_b) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c
```

```python
from sklearn.neighbors import KNeighborsClassifier

df_dots = df_dataset[['lat','lng']].dropna()

knc = KNeighborsClassifier(metric=distance)
knc.fit(X=df_dots , y=np.ones(df_dots.shape[0]))
distances, indexes = knc.kneighbors(X=df_dots, n_neighbors=6,)
```

```python
for i in range(1, 6):
    df_dots[f'distance_{i}'] = distances[:,i]
    df_dots[f'indexes_{i}'] = indexes[:,i]
    # df_dots[f'mean_{i}'] = distances[:,i].mean()
    # df_dots[f'var_{i}'] = distances[:,i].var()
    
df_dots['mean'] = df_dots.iloc[:,df_dots.columns.str.contains('distance')].mean(axis=1)
```

### Категории банкоматов в окрестности

```python
from sklearn.neighbors import DistanceMetric

R = 6373.0 # радиус Земли в километрах

coords = np.deg2rad(df_dataset[['lat', 'lng']])
metric = DistanceMetric.get_metric('haversine')
dist = metric.pairwise(coords.values) * R
```

```python
from functools import reduce
from collections import defaultdict, Counter

def get_closest_atm_groups(dist, radius):
    mask = np.logical_and(0 < dist, dist <= radius)
    index = reduce(lambda d, p: d[p[0]].append(p[1]) or d, zip(*np.where(mask)), defaultdict(list))
    df_atms_groups = {
        df_dataset.loc[i, 'id'] : Counter([str(df_dataset.loc[j, 'atm_group']) for j in i_list])
        for i, i_list in index.items()
    }
    
    df_atm_groups = [dict({'id': i}, **group) for i, group in df_atms_groups.items()]
    df_atm_groups = pd.DataFrame(df_atm_groups)
    df_atm_groups.columns = list(map(lambda s: f'atm_group_{s}_{radius}km', df_atm_groups.columns[:-1])) + \
            [df_atm_groups.columns[-1]]
    df_atm_groups.fillna(0, inplace=True)
    df_atm_groups = df_atm_groups.astype(int, copy=False)
    
    return df_atm_groups
```

```python
df_atm_groups = get_closest_atm_groups(dist, 0.5)
df_dataset = pd.merge(df_dataset, df_atm_groups, how='left', on='id')

columns = df_atm_groups.columns[df_atm_groups.columns.str.startswith('atm_group_')]
df_dataset[columns.tolist()] = df_dataset[columns.tolist()].fillna(0)

# df_atm_groups = get_closest_atm_groups(dist, 0.7)
# df_dataset = pd.merge(df_dataset, df_atm_groups, how='left', on='id')

# columns = df_atm_groups.columns[df_atm_groups.columns.str.startswith('atm_group_')]
# df_dataset[columns.tolist()] = df_dataset[columns.tolist()].fillna(0)

df_atm_groups = get_closest_atm_groups(dist, 1.2)
df_dataset = pd.merge(df_dataset, df_atm_groups, how='left', on='id')

columns = df_atm_groups.columns[df_atm_groups.columns.str.startswith('atm_group_')]
df_dataset[columns.tolist()] = df_dataset[columns.tolist()].fillna(0)
```

### Средние тагреты

```python
from sklearn.model_selection import KFold

def count_mean_target(df, cols, alpha=0.2, n_folds=5, random_state=4896):
    colname = 'mean_target_' + '_'.join(cols)
    
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=random_state)
    unique = df.groupby(cols)['target'].count().reset_index()[cols]
    
    mean_target = lambda r: (r['mean'] * r['count'] + alpha * r['target']) / (r['count'] + alpha)
    
    for train_index, valid_index in kf.split(df):
        mean_g = df.loc[train_index, 'target'].mean()
        
        mean_r = df.loc[train_index].groupby(cols)['target'].agg(['mean', 'count']).reset_index()
        mean_r = pd.merge(unique, mean_r, on=cols, how='left').fillna(0.0)
        mean_r['target'] = mean_g
        mean_r['target'] = mean_r.apply(mean_target, axis=1)
        mean_r.drop(columns=['mean', 'count'], inplace=True)
    
        df.loc[valid_index, colname] = pd.merge(df.loc[valid_index, cols], mean_r, on=cols, how='left')['target']
        
    return df
```
