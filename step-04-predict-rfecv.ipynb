```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from IPython.display import clear_output
```

```python
df_dataset = pd.read_csv('dataset_features.csv', sep=',')
df_dataset.head()
```

```python
mask = df_dataset['target'].notna()
df_train = df_dataset.loc[mask]
print(df_train.shape)
df_train.head()
```

```python
mask = np.logical_not(mask)
df_test = df_dataset.loc[mask]
print(df_test.shape)
df_test.head()
```

```python
features = np.asarray(df_dataset.columns.tolist()[:-1])
target = df_dataset.columns.tolist()[-1]
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
cl_params = {}
```

## XGBoost

```python
import xgboost as xgb
```

```python
from xgboost import XGBRegressor
```

```python
%%time

params = {
    'max_depth': 5,
    'learning_rate': 0.01,
    'silent': 0,
    'objective': 'reg:linear',
    'seed': 8888,
}

X_train = xgb.DMatrix(df_train.loc[:, features], label=df_train.loc[:, target])

trees = 2000
cv = xgb.cv(params, X_train, metrics=('rmse'), show_stdv=False,
            num_boost_round=trees, nfold=5, early_stopping_rounds=15)
```

```python
cv[100:].plot(figsize=(6, 6), y=['test-rmse-mean', 'train-rmse-mean'])
print(cv.loc[cv['test-rmse-mean'].values.argmin()])
trees = cv['test-rmse-mean'].values.argmin()
trees
```

```python
params['n_estimators'] = trees
cl_params['xgb'] = params
```

## LightGBM

```python
import lightgbm as lgbm
```

```python
from lightgbm import LGBMRegressor
```

```python
cl = LGBMRegressor(
    objective='regression', 
    max_depth=5,
    colsample_bytre=0.8,
    subsample=0.8,
    learning_rate=0.005,
    n_estimators=3000
)
```

```python
%%time

from sklearn.feature_selection import RFECV

rfecv = RFECV(estimator=cl, step=10, cv=5, scoring='neg_mean_squared_error', verbose=1)
rfecv.fit(df_train.loc[:, features].values, df_train.loc[:, target].values)

clear_output()
```

```python
plt.figure(figsize=(6, 6))
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

```python
%%time

params = {
    'max_depth': 5,
    'learning_rate': 0.002,
    'colsample_bytre': 0.8,
    'subsample': 0.8,
    'silent': 0,
    'objective': 'regression',
    'seed': 8888,
}

X_train = lgbm.Dataset(rfecv.transform(df_train.loc[:, features]), label=df_train.loc[:, target])

trees = 12000
cv = lgbm.cv(params, X_train, metrics=('rmse'), show_stdv=False, stratified=False,
             num_boost_round=trees, nfold=5, early_stopping_rounds=30)
```

```python
cv = pd.DataFrame(cv)
```

```python
cv[100:].plot(figsize=(6, 6), y=['rmse-mean'])
print(cv.loc[cv['rmse-mean'].values.argmin()])
trees = cv['rmse-mean'].values.argmin()
trees
```

```python
params['n_estimators'] = trees
cl_params['lgb'] = params
```

```python
cl_params['lgb']
```

```python
%%time

scores = []
feature_importances = []

kf = KFold(n_splits=5, shuffle=True, random_state=8888)

for train_index, valid_index in kf.split(df_train):
    cl = LGBMRegressor(**cl_params['lgb'], random_state=8888)
    
    X_train = rfecv.transform(df_train.iloc[train_index].loc[:, features].values)
    Y_train = df_train.iloc[train_index].loc[:, target].values
    
    X_valid = rfecv.transform(df_train.iloc[valid_index].loc[:, features].values)
    Y_valid = df_train.iloc[valid_index].loc[:, target].values
    
    cl.fit(X_train, Y_train)
    Y_pred = cl.predict(X_valid)
    
    feature_importances.append(cl.feature_importances_)
    scores.append(np.sqrt(mean_squared_error(Y_valid, Y_pred)))
    print(scores[-1])
    
print("=" * 30)
print("MEAN:", np.mean(scores))
```

```python
features_ = np.array(features)

importances = np.mean(feature_importances, axis=0)
importances = importances / importances.sum()
indices = np.argsort(importances)[::-1]

print("Feature ranking:")

n_features = len(features_)
n_cut = 100

for f in range(n_cut):
    print("[{:03}] {:50} {}".format(f + 1, features_[indices[f]], importances[indices[f]]))
    
plt.figure(figsize=(16, 6))
plt.title("Feature importances")
plt.bar(range(n_cut), importances[indices][:n_cut], color="r", align="center")
plt.xticks(range(n_cut), features_[indices][:n_cut], rotation='vertical')
plt.xlim([-1, n_cut])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_test = df_dataset.loc[df_dataset['target'].isna()]

tmp = pd.read_csv('../test.csv', sep=',', index_col=0)
tmp = pd.DataFrame(tmp['id'].astype(int))
tmp = pd.merge(tmp, df_test, on='id', how='left')

df_test = tmp

X_test = rfecv.transform(df_test.loc[:, features].values)
```

```python
scores, Y_pred_test = [], []

for seed in np.random.randint(100, 10000, 10):    
    X_train, X_valid, Y_train, Y_valid = train_test_split(
        df_train.loc[:, features].values,
        df_train.loc[:, target].values,
        test_size=0.2, random_state=seed)
    
    X_train = rfecv.transform(X_train)
    X_valid = rfecv.transform(X_valid)
    
    cl = LGBMRegressor(**cl_params['lgb'])
    cl.fit(X_train, Y_train)
    Y_pred = cl.predict(X_valid)
    scores.append(np.sqrt(mean_squared_error(Y_valid, Y_pred)))
    print("seed={:4d}; rmse={}".format(seed, scores[-1]))
    
    Y_pred = cl.predict(X_test)
    Y_pred_test.append(Y_pred)
    
print("=" * 30)
print("MEAN:", np.mean(scores))
```

```python
Y_pred = np.mean(Y_pred_test, axis=0)
assert len(Y_pred) == df_test.shape[0]
```

```python
df_pred = pd.DataFrame(Y_pred, columns=['target'])
df_pred.to_csv('submission_11.csv', index=True, sep=',')
```
